```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

## function that return a list of words in descending frequency, each tuple = (word, # of occurrence)
def top_n_words(corpus,n):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0,idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x:x[1], reverse=True)
    return words_freq[:n]

## the two feature sets for yelp and IMDB
ye_feature = top_n_words(ye_train_f,10000)   ## top 10,000 words in yelp, with frequency in each tuple
im_feature = top_n_words(im_train_f,10000)   ## top 10,000 words in IMDB, with frequency in each tuple

vocab_ye = []
for t in ye_feature:
    vocab_ye.append(t[0])
vocab_im = []
for t in im_feature:
    vocab_im.append(t[0])
```

```python
## function to calculate frequency bag of words
def freq(arr):
    array = []
    for i in arr:
        s = i.sum()
        if s == 0:
            array.append(np.zeros(arr.shape[1]))
        else:
            array.append(i.astype(float)/s)
    return np.asarray(array)

## now creaet the Frequency BoW representation, as np arrays.
freq_ye_train = freq(vec_ye_train)
freq_ye_valid = freq(vec_ye_valid)
freq_ye_test = freq(vec_ye_test)
freq_im_train = freq(vec_im_train)
freq_im_valid = freq(vec_im_valid)
freq_im_test = freq(vec_im_test)

##print(freq_ye_train.shape)
##aa = np.array([[1,2,3],[0,0,0]])
##print(freq(aa))
```

```python
## function that converts a review to ids
def conv(strr,voc):
    arr = []
    for s in strr.split():
        if s in voc:
            arr.append(voc.index(s))
    str1 = ' '.join(str(e) for e in arr)
    return str1

## create the id encoded datapoints  - This chunk takes around 40 minutes.
ye_train_f_num = []
for i in ye_train_f:
    ye_train_f_num.append(conv(i,words_ye))
ye_valid_f_num = []
for i in ye_valid_f:
    ye_valid_f_num.append(conv(i,words_ye))
ye_test_f_num = []
for i in ye_test_f:
    ye_test_f_num.append(conv(i,words_ye))
im_train_f_num = []
for i in im_train_f:
    im_train_f_num.append(conv(i,words_im))
im_valid_f_num = []
for i in im_valid_f:
    im_valid_f_num.append(conv(i,words_im))
im_test_f_num = []
for i in im_test_f:
    im_test_f_num.append(conv(i,words_im))


##list1 = ["foo","bar","baz"]
##st = "foo baz bar ba"
##print(conv(st,list1))
```
